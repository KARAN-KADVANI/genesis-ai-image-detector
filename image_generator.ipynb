```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import os
import numpy as np

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)
```

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to(device)

# Disable safety checker to avoid black images
pipe.safety_checker = None
```

```python
prompts = [
    "Highly realistic photo of a modern industrial cleanroom, stainless steel machinery, bright fluorescent lighting, ultra wide angle, DSLR photography",
    
    "Highly realistic image of an automated factory floor with conveyor belts and robotic arms, sharp focus, industrial photography, realistic reflections",
    
    "Highly realistic photo of a semiconductor manufacturing facility, sterile white environment, professional industrial photography",
    
    "Highly realistic image of a high-tech laboratory interior, scientific equipment, clean environment, natural perspective, DSLR photo",
    
    "Highly realistic photo of an empty industrial production line, symmetrical composition, overhead lighting, professional camera",
    
    "Highly realistic image of a large manufacturing plant interior, depth of field, realistic textures, industrial realism"

    "Highly realistic photo of a rocket standing on a launch pad, daytime lighting, sharp details, professional aerospace photography",
    
    "Highly realistic image of a space launch facility with service towers, realistic metal textures, DSLR photography",
    
    "Highly realistic photo of a rocket engine test facility, industrial setting, dramatic lighting, ultra detailed",
    
    "Highly realistic image of aerospace infrastructure, large mechanical structures, realistic shadows, professional photo",
    
    "Highly realistic photo of a spacecraft assembly building interior, wide angle lens, natural colors",
    
    "Highly realistic image of a launch pad with structural scaffolding, realistic perspective, documentary style"

    "Highly realistic photo of a modern office building interior, natural daylight, clean design, DSLR photography",
    
    "Highly realistic image of an empty airport terminal, symmetrical architecture, realistic lighting",
    
    "Highly realistic photo of a large underground metro station, wide angle lens, realistic shadows",
    
    "Highly realistic image of a futuristic research facility interior, minimal design, professional photography",
    
    "Highly realistic photo of an industrial warehouse interior, depth of field, realistic textures"

    "Highly realistic photo of a data center corridor with server racks, LED lighting, professional photography",
    
    "Highly realistic image of a power plant interior, industrial equipment, realistic metallic textures",
    
    "Highly realistic photo of a control room with large screens and panels, sharp focus, documentary photography",
    
    "Highly realistic image of a research facility hallway, modern design, DSLR photo",
    
    "Highly realistic photo of an empty technology lab, natural lighting, realistic camera noise"

    "A realistic photo of a study table with books and a laptop"
    "A realistic photo of a ceiling fan in a room"
    "A realistic photo of a classroom blackboard with chalk dust"
    "A realistic photo of a wall clock hanging in an office"
    "A realistic photo of a shelf filled with files and folders"

]

def is_black_image(image, threshold=10):
    img = np.array(image)
    return img.mean() < threshold
```

```python
save_dir = r"C:\Users\ASUS\Desktop\AI_Image_Detector\final_dataset\train\AI"
os.makedirs(save_dir, exist_ok=True)

existing = [
    int(f.split("_")[1].split(".")[0])
    for f in os.listdir(save_dir)
    if f.startswith("ai_")
]

start_index = max(existing) + 1 if existing else 0

num_images_to_generate = 1000   # generate 1000 NEW images
end_index = start_index + num_images_to_generate

print(f"Starting from index: {start_index}")
print(f"Will generate till index: {end_index - 1}")

i = start_index
while i < end_index:
    prompt = prompts[i % len(prompts)]

    generator = torch.Generator(device).manual_seed(
        torch.randint(0, 1_000_000, (1,)).item()
    )

    image = pipe(
        prompt,
        guidance_scale=7.0,
        num_inference_steps=25,
        generator=generator
    ).images[0]

    if is_black_image(image):
        print("⚠️ Black image skipped")
        continue

    image.save(f"{save_dir}/ai_{i}.png")

    if i % 50 == 0:
        print(f"✅ Saved up to image index {i}")

    i += 1
```
